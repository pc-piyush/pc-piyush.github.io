# Welcome

I am a Data Analyst working at [UF Health Integrated Data Repository](https://idr.ufhealth.org/). My work primarily focuses on designing and developing data science pipelines to improve the coverage and quality of the data, primarily for the [OneFlorida+ Clinical Research Consortium](https://onefloridaconsortium.org/). In my current role, I work on common data models like [OMOP](https://ohdsi.github.io/CommonDataModel/) and [PCORNET](https://pcornet.org/news/resources-pcornet-common-data-model/). Additionally, I also help research at UF by providing consultation for research teams, supporting feasibility and cohort discovery, and receiving and fulfilling research data requests.
